```python
import pandas as pd
df = pd.read_csv("data/weather-missing.csv", parse_dates=["day"])
df
```

```python
df2 = df.fillna(0)
df2
```

```python
df2 = df.fillna({
    "temperature": 0,
    "windspeed": 0,
    "event": "not-available"
})
df2
```

```python
df2 = df.fillna(method="ffill")
df2
```

```python
df2 = df.fillna(method="ffill", limit=1)
df2
```

```python
df2 = df.fillna(method="bfill")
df2
```

```python
df2 = df.fillna(method="bfill", axis="columns")
df2
```

```python
df2 = df.dropna()
df2
```

```python
df2 = df.set_index("day")
df2 = df2.dropna(how="all")
df2
```

```python
df2 = df2.dropna(thresh=1)
df2
```
