```python
import pandas as pd
df = pd.read_csv("data/weather-missing.csv", parse_dates=["day"])
df
```

```python
df2 = df.fillna({
    "temperature": 1000,
    "windspeed": 500,
    "event": "not-available"
})
df2
```

```python
import numpy as np
df2.replace(1000, np.NaN)
```

```python
import numpy as np
df2.replace([1000, 500, "not-available"], np.NaN)
```

```python
df2.replace({"temperature":1000, "event":"not-available"}, np.NaN)
```

```python
df2.replace({1000:np.NaN,"not-available":"Unavailable"})
```

```python
df2.replace({
    "temperature": "[A-Za-z]",
    "windspeed": "[A-Za-z]"
}, "", regex=True)
```
