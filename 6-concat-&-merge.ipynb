# merge

```python
df1 = pd.DataFrame({
    "city": ["Mumbai", "Delhi", "Bangalore"],
    "temperature": [32,45,20],
    "humidity": [80,60,70]
})
df2 = pd.DataFrame({
    "city": ["Mumbai", "Delhi", "Bangalore"],
    "windspeed": [7,12,9]
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2, on="city")
```

```python
df1 = pd.DataFrame({
    "city": ["Mumbai", "Delhi", "Chennai"],
    "temperature": [32,45,20],
    "humidity": [80,60,70]
})
df2 = pd.DataFrame({
    "city": ["Mumbai", "Delhi", "Kolkata"],
    "windspeed": [7,12,9]
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2, on="city")
```

```python
pd.merge(df1,df2, on="city", how="outer")
```

```python
pd.merge(df1,df2, on="city", how="left")
```

```python
pd.merge(df1,df2, on="city", how="right")
```

```python
pd.merge(df1,df2, on="city", how="outer", indicator=True)
```
